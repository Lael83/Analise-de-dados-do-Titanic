```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Lendo a base de dados
base = pd.read_csv("Titanic-Dataset.csv")
base.columns = ["IDpassageiro","Sobrevivente","ClassePassagem","Nome","Genero","Idade","NrIrmaosConjuges",
                "NrPaisFilhos","NrTicket","Tarifa","Cabine","PortoEmbarque"]

# Contagem de passageiros por gênero
genero_count = base['Genero'].value_counts()

print("Número de Passageiros por Gênero:")
print(genero_count)

# Gráfico de barras
plt.figure(figsize=(8, 6))
sns.barplot(x=genero_count.index, y=genero_count.values, palette='viridis')
plt.title('Distribuição de Gênero dos Passageiros')
plt.xlabel('Gênero')
plt.ylabel('Número de Passageiros')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(base['Idade'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Distribuição de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Número de Passageiros')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Lendo a base de dados
base = pd.read_csv("Titanic-Dataset.csv")
base.columns = ["IDpassageiro", "Sobrevivente", "ClassePassagem", "Nome", "Genero", "Idade", "NrIrmaosConjuges",
                "NrPaisFilhos", "NrTicket", "Tarifa", "Cabine", "PortoEmbarque"]

# Calculando a quantidade de sobreviventes e não sobreviventes
quantidade_sobreviventes = base['Sobrevivente'].value_counts()

# Criando um gráfico de barras para a quantidade de sobreviventes
plt.figure(figsize=(6, 4))
sns.barplot(x=quantidade_sobreviventes.index, y=quantidade_sobreviventes.values, palette='viridis')
plt.title('Quantidade de Sobreviventes')
plt.xlabel('Sobrevivente')
plt.ylabel('Número de Passageiros')
plt.show()

# Explorando a quantidade de sobreviventes por classe
quantidade_sobreviventes_classe = base.groupby('ClassePassagem')['Sobrevivente'].sum()

# Criando um gráfico de barras para a quantidade de sobreviventes por classe
plt.figure(figsize=(8, 6))
sns.barplot(x=quantidade_sobreviventes_classe.index, y=quantidade_sobreviventes_classe.values, palette='viridis')
plt.title('Quantidade de Sobreviventes por Classe')
plt.xlabel('Classe de Passagem')
plt.ylabel('Número de Sobreviventes')
plt.show()

# Explorando a quantidade de sobreviventes por gênero
quantidade_sobreviventes_genero = base.groupby('Genero')['Sobrevivente'].sum()

# Criando um gráfico de barras para a quantidade de sobreviventes por gênero
plt.figure(figsize=(6, 4))
sns.barplot(x=quantidade_sobreviventes_genero.index, y=quantidade_sobreviventes_genero.values, palette='viridis')
plt.title('Quantidade de Sobreviventes por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Número de Sobreviventes')
plt.show()

# Explorando a quantidade de sobreviventes por faixa etária
base['FaixaEtaria'] = pd.cut(base['Idade'], bins=[0, 18, 35, 50, 100], labels=['0-18', '19-35', '36-50', '51+'])
quantidade_sobreviventes_idade = base.groupby('FaixaEtaria')['Sobrevivente'].sum()

# Criando um gráfico de barras para a quantidade de sobreviventes por faixa etária
plt.figure(figsize=(8, 6))
sns.barplot(x=quantidade_sobreviventes_idade.index, y=quantidade_sobreviventes_idade.values, palette='viridis')
plt.title('Quantidade de Sobreviventes por Faixa Etária')
plt.xlabel('Faixa Etária')
plt.ylabel('Número de Sobreviventes')
plt.show()
```

```python
import pandas as pd

# Lendo a base de dados
base = pd.read_csv("Titanic-Dataset.csv")
base.columns = ["IDpassageiro", "Sobrevivente", "ClassePassagem", "Nome", "Genero", "Idade", "NrIrmaosConjuges",
                "NrPaisFilhos", "NrTicket", "Tarifa", "Cabine", "PortoEmbarque"]

# Filtrando os sobreviventes
sobreviventes = base[base['Sobrevivente'] == 1]

# Calculando a distribuição do gênero dos sobreviventes
distribuicao_genero_sobreviventes = sobreviventes['Genero'].value_counts().reset_index()
distribuicao_genero_sobreviventes.columns = ['Gênero', 'Número de Sobreviventes']

# Exibindo a tabela
print("Distribuição do Gênero dos Sobreviventes:")
print(distribuicao_genero_sobreviventes)

```
